```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def n_Sel(lambda_,a0,aUV,aIR):
    return np.sqrt(a0+aUV*lambda_**2/(lambda_**2-146.9**2)+aIR*lambda_**2/(lambda_**2-827.0**2))

a0_offset=0.0
def n_Sel_178K(lambda_):
    return n_Sel(lambda_,1.4+a0_offset,0.4,0.002)

def n_Sel_162K(lambda_):
    return n_Sel(lambda_,1.5+a0_offset,0.38,0.009)
```

```python
# lambda_list=np.linspace(160,200,41)
lambda_list=np.linspace(175,182,41)
n_178K_list=n_Sel_178K(lambda_list)
n_162K_list=n_Sel_162K(lambda_list)
plt.plot(lambda_list,n_178K_list,label="178 K")
plt.plot(lambda_list,n_162K_list,label="162 K")
plt.gca().axvline(x=178,color="r")
plt.xlabel("Wavelength (nm)")
plt.ylabel("n_LXe")
plt.legend()
plt.show()
# Want index at 178 nm and triple point (162 K) to match measured value of 1.69
# ... that's what Grace paper says, but Solovov found 1.69 +/- 0.02 at 170 K, not 162 K
# The values listed in the paper don't seem to produce the same curve...
print(n_Sel_162K(175),n_Sel_162K(178),n_Sel_162K(180))
print(n_Sel_178K(175),n_Sel_178K(178),n_Sel_178K(180))
print(n_Sel_162K(361.2),n_Sel_162K(470.0),n_Sel_162K(643.9))
```

```python
# From Yang 2008 paper (DuPont)
def n_Sel_test(lambda_, a, b1, b2, c1, c2):
    return np.sqrt(a+b1*lambda_**2/(lambda_**2-c1**2)+b2*lambda_**2/(lambda_**2-c2**2))

# Here, absorption edge for Teflon AF (a copolymer of TFE and PDD) is set to ~126 nm; 
# gives a good fit to data >160 nm or so, but doesn't match absorption peak position ~155 nm
lambda_test = np.linspace(150,200,1000)
n_Sel_plot = n_Sel_test(lambda_test/1000, 1.517, .184, 1, np.sqrt(0.016), np.sqrt(104.66)) # wavelengths used in paper are in um
plt.plot(lambda_test, n_Sel_plot)
```
