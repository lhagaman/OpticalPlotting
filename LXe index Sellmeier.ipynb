```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def n_Sel(lambda_,a0,aUV,aIR):
    return np.sqrt(a0+aUV*lambda_**2/(lambda_**2-146.9**2)+aIR*lambda_**2/(lambda_**2-827.0**2))

def n_Sel_178K(lambda_):
    return n_Sel(lambda_,1.4,0.4,0.002)

def n_Sel_162K(lambda_):
    return n_Sel(lambda_,1.5,0.38,0.009)
```

```python
# lambda_list=np.linspace(160,200,41)
lambda_list=np.linspace(175,182,41)
n_178K_list=n_Sel_178K(lambda_list)
n_162K_list=n_Sel_162K(lambda_list)
plt.plot(lambda_list,n_178K_list,label="178 K")
plt.plot(lambda_list,n_162K_list,label="162 K")
plt.gca().axvline(x=178,color="r")
plt.xlabel("Wavelength (nm)")
plt.ylabel("n_LXe")
plt.legend()
plt.show()
print(n_Sel_162K(175),n_Sel_162K(178),n_Sel_162K(180))
```
